```python
from pytrends.request import TrendReq
import pandas as pd

pytrends = TrendReq(hl='en-US', tz=360)
kw_list = ["moisturizer","toner","mask","perfume","blush"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 3-m', geo='US', gprop='')
```

```python
related_queries = pytrends.related_queries()
df = pd.DataFrame(columns=["product_name", "hit_count", "product_type"])
for keyword in kw_list:
  df_product = related_queries[keyword]['rising'].head(10)
  df_product.columns = ["product_name", "hit_count"]
  df_product["product_type"] = keyword
  # Concatenate them
  df = pd.concat([df, df_product], ignore_index=True)
print(df)
```

```python
df.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("workspace.default.google_trend_items")
```
